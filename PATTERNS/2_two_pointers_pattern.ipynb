## 1. Définitions et concepts clés

- **Two Pointers Pattern** : Technique algorithmique utilisant deux pointeurs pour parcourir une structure de données, généralement dans des directions opposées ou à des vitesses différentes [1](#ref1).
- **Pointeur** : Variable utilisée pour stocker l'adresse mémoire d'une autre variable ou d'un élément dans une structure de données [2](#ref2).
- **Complexité spatiale** : Mesure de la quantité de mémoire utilisée par un algorithme [1](#ref1).
- **Complexité temporelle** : Mesure du temps d'exécution d'un algorithme en fonction de la taille de l'entrée [1](#ref1).
- **In-place algorithm** : Algorithme qui modifie directement l'entrée sans utiliser de structure de données supplémentaire significative [3](#ref3).

Le Two Pointers Pattern est une technique algorithmique puissante qui permet de résoudre efficacement de nombreux problèmes, en particulier ceux impliquant des tableaux triés ou des chaînes de caractères. Cette approche utilise deux pointeurs pour parcourir une structure de données, souvent en se déplaçant dans des directions opposées ou à des vitesses différentes, ce qui permet de réduire la complexité temporelle de nombreux algorithmes de O(n^2) à O(n).

## 2. Histoire et introduction au Two Pointers Pattern

Le Two Pointers Pattern, bien que non formellement défini comme une technique distincte jusqu'à récemment, a des racines profondes dans l'histoire de l'informatique et de l'algorithmique. Son utilisation implicite remonte aux premiers jours de la programmation, où l'optimisation de la mémoire et du temps d'exécution était cruciale en raison des ressources limitées des ordinateurs.

L'un des premiers exemples notables d'utilisation de cette technique peut être trouvé dans l'algorithme de recherche binaire, développé dans les années 1940 [4](#ref4). Bien que la recherche binaire n'utilise qu'un seul pointeur mobile, elle illustre le principe fondamental de réduire l'espace de recherche de manière efficace, qui est au cœur du Two Pointers Pattern.

Dans les années 1970 et 1980, avec l'émergence de problèmes algorithmiques plus complexes et la nécessité d'optimiser les performances, la technique des deux pointeurs a commencé à être utilisée plus explicitement. Un exemple célèbre est l'algorithme de partition de Hoare, utilisé dans le tri rapide (Quicksort), inventé par Tony Hoare en 1959 [5](#ref5). Cet algorithme utilise deux pointeurs se déplaçant dans des directions opposées pour partitionner efficacement un tableau.

Cependant, c'est vraiment avec l'essor de la programmation compétitive et des entretiens techniques dans l'industrie du logiciel que le Two Pointers Pattern a été reconnu et nommé comme une technique distincte. Des problèmes classiques comme "Trouver deux nombres dans un tableau trié dont la somme est égale à une valeur cible" ont popularisé cette approche.

Aujourd'hui, le Two Pointers Pattern est enseigné comme une technique fondamentale en algorithmique et est largement utilisé dans divers domaines de l'informatique, de l'optimisation des requêtes de base de données à l'analyse de séquences en bio-informatique.

La puissance du Two Pointers Pattern réside dans sa capacité à :
1. Réduire la complexité temporelle de nombreux algorithmes.
2. Optimiser l'utilisation de la mémoire en travaillant in-place.
3. Simplifier la logique de nombreux algorithmes, les rendant plus faciles à comprendre et à maintenir.

Dans les sections suivantes, nous explorerons en détail comment implémenter et appliquer cette technique puissante à divers problèmes.

## 3. Implémentation et cas d'utilisation

Le Two Pointers Pattern est particulièrement efficace pour résoudre des problèmes impliquant des séquences linéaires, comme les tableaux ou les chaînes de caractères. Pour mieux comprendre le fonctionnement du Two Pointers Pattern, nous allons implémenter l'algorithme classique de recherche de deux nombres dans un tableau trié dont la somme est égale à une valeur cible. Ensuite, nous visualiserons son fonctionnement de deux manières différentes.

### 3.1 Implémentation de base

```python
# type: ignore
# Third-party library imports
from IPython.display import display, HTML
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def two_sum(arr, target):
    """
    Trouve deux nombres dans un tableau trié dont la somme est égale à la cible.
    
    Args:
    arr (List[int]): Un tableau trié d'entiers.
    target (int): La somme cible.
    
    Returns:
    Tuple[int, int]: Les indices des deux nombres, ou None si non trouvés.
    """
    left, right = 0, len(arr) - 1
    while left < right:
        current_sum = arr[left] + arr[right]
        if current_sum == target:
            return (left, right)
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    return None  # Aucune solution trouvée

# Exemple d'utilisation
arr = [1, 3, 4, 5, 7, 10, 11]
target = 9
result = two_sum(arr, target)
print(f"Indices des nombres dont la somme est {target}: {result}")
```

Cette implémentation utilise deux pointeurs, `left` et `right`, qui se déplacent l'un vers l'autre jusqu'à trouver la paire de nombres dont la somme est égale à la cible ou jusqu'à ce qu'ils se croisent.

### 3.2 Visualisation étape par étape

```python
def visualize_two_pointers(arr, target):
    left, right = 0, len(arr) - 1
    data = []
    found = False
    
    while left < right:
        current_sum = arr[left] + arr[right]
        step = {
            'Étape': len(data) + 1,
            'Pointeur gauche': left,
            'Pointeur droit': right,
            'Valeur gauche': arr[left],
            'Valeur droite': arr[right],
            'Somme': current_sum,
            'Action': ''
        }
        
        if current_sum == target:
            step['Action'] = 'Trouvé ! Terminé.'
            data.append(step)
            found = True
            break
        elif current_sum < target:
            step['Action'] = 'Somme trop petite, déplacer pointeur gauche'
            left += 1
        else:
            step['Action'] = 'Somme trop grande, déplacer pointeur droit'
            right -= 1
        
        data.append(step)
    
    if not found:
        data.append({
            'Étape': len(data) + 1,
            'Pointeur gauche': left,
            'Pointeur droit': right,
            'Valeur gauche': '-',
            'Valeur droite': '-',
            'Somme': '-',
            'Action': 'Aucune solution trouvée'
        })
    
    df = pd.DataFrame(data)
    
    styled_df = df.style.set_properties(**{
        'text-align': 'center',
        'font-size': '14px',
        'border-color': 'black',
        'border-style': 'solid',
        'border-width': '1px'
    }).set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#f0f0f0'), ('font-weight', 'bold')]},
        {'selector': 'td', 'props': [('padding', '10px')]},
        {'selector': '', 'props': [('border-collapse', 'collapse')]}
    ])
    
    display(HTML(f"<h3>Recherche de deux nombres dont la somme est égale à {target}</h3>"))
    display(HTML(f"<p>Tableau trié : {arr}</p>"))
    display(styled_df)

# Exemple d'utilisation
arr = [1, 3, 4, 5, 7, 10, 11]
target = 9
visualize_two_pointers(arr, target)
```

### 3.3 Visualisation graphique

```python
def visualize_two_pointers_graph(arr, target):
    left, right = 0, len(arr) - 1
    steps = []
    found = False
    
    while left < right:
        current_sum = arr[left] + arr[right]
        steps.append((left, right, current_sum))
        
        if current_sum == target:
            found = True
            break
        elif current_sum < target:
            left += 1
        else:
            right -= 1
    
    fig, ax = plt.subplots(figsize=(14, 10))
    ax.set_title(f"Recherche de deux nombres dont la somme est égale à {target}", fontsize=16)
    ax.set_xlabel("Index", fontsize=12)
    ax.set_ylabel("Valeur", fontsize=12)
    
    # Tracer le tableau
    ax.scatter(range(len(arr)), arr, c='blue', s=100)
    for i, v in enumerate(arr):
        ax.text(i, v, str(v), ha='center', va='bottom', fontsize=10)
    
    # Tracer les mouvements des pointeurs
    max_value = max(arr)
    for i, (left, right, sum_val) in enumerate(steps):
        color = 'green' if sum_val == target else 'red'
        
        # Calculer la position verticale de la flèche
        arrow_y = max_value + 2 + (i * 3)
        
        # Tracer la flèche
        ax.annotate('', xy=(left, arrow_y), xytext=(right, arrow_y),
                    arrowprops=dict(arrowstyle='<->', color=color, lw=2))
        
        # Ajouter le texte au-dessus de la flèche
        ax.text((left + right) / 2, arrow_y + 0.5, f'Étape {i+1}: Somme = {sum_val}',
                ha='center', va='bottom', fontsize=10, 
                bbox=dict(boxstyle='round,pad=0.5', fc='yellow', ec='orange', alpha=0.8))
        
        # Ajouter des lignes pointillées reliant les points aux flèches
        ax.plot([left, left], [arr[left], arrow_y], 'k--', alpha=0.5)
        ax.plot([right, right], [arr[right], arrow_y], 'k--', alpha=0.5)

    plt.ylim(0, max_value + len(steps) * 3.5)
    plt.tight_layout()
    plt.show()

# Exemple d'utilisation
arr = [1, 3, 4, 5, 7, 10, 11]
target = 9
visualize_two_pointers_graph(arr, target)
```

### 3.4 Autres cas d'utilisation:

#### 3.4.1 Recherche de paires dans un tableau trié

Comme illustré dans l'exemple précédent, cette technique est idéale pour trouver des paires d'éléments dans un tableau trié qui satisfont une certaine condition (par exemple, une somme spécifique).

#### 3.4.2 Inversion de chaînes ou de tableaux

Le Two Pointers Pattern peut être utilisé pour inverser efficacement une chaîne ou un tableau en place, en utilisant deux pointeurs qui se déplacent l'un vers l'autre depuis les extrémités (exemple ci-dessous).

#### 3.4.3 Suppression de doublons

Dans un tableau trié, on peut utiliser deux pointeurs pour supprimer les doublons efficacement, en maintenant un pointeur pour les éléments uniques et un autre pour parcourir le tableau.

#### 3.4.4 Problèmes de fenêtre coulissante

Bien que techniquement une variante, le pattern de fenêtre coulissante utilise souvent deux pointeurs pour définir les limites d'une sous-séquence d'intérêt.

#### 3.4.5  Détection de cycles dans une liste chaînée

L'algorithme du lièvre et de la tortue, qui utilise deux pointeurs se déplaçant à des vitesses différentes, est un exemple classique d'utilisation du Two Pointers Pattern pour détecter des cycles 

```python
def reverse_array(arr):
    """
    Inverse un tableau en place en utilisant le Two Pointers Pattern.
    
    Args:
    arr (List): Le tableau à inverser.
    
    Returns:
    List: Le tableau inversé (même objet, modifié en place).
    """
    left = 0
    right = len(arr) - 1
    
    while left < right:
        # Échange les éléments aux positions left et right
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    
    return arr

# Exemple d'utilisation
original_array = [1, 2, 3, 4, 5]
reversed_array = reverse_array(original_array)
print(f"Tableau original: {original_array}")
print(f"Tableau inversé: {reversed_array}")
```

Le Two Pointers Pattern trouve de nombreuses applications dans des scénarios réels de programmation et de résolution de problèmes. Voici quelques exemples détaillés :

### 4.1 Traitement de texte et manipulation de chaînes

Dans le traitement de texte, le Two Pointers Pattern est souvent utilisé pour des tâches telles que la vérification des palindromes ou la compression de chaînes.

#### Exemple : Vérification de palindrome

```python
def is_palindrome(s):
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True

# Test
print(is_palindrome("radar"))  # True
print(is_palindrome("hello"))  # False
```

### 4.4 Algorithmes de tri

Le Two Pointers Pattern est utilisé dans plusieurs algorithmes de tri, comme le tri rapide (Quicksort) pour le partitionnement du tableau.

#### Exemple : Partition de Lomuto (utilisée dans Quicksort)

```python
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

# Utilisation dans Quicksort
def quicksort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quicksort(arr, low, pi - 1)
        quicksort(arr, pi + 1, high)

# Test
arr = [10, 7, 8, 9, 1, 5]
quicksort(arr, 0, len(arr) - 1)
print("Tableau trié:", arr)
```

## 5. Comparaison avec d'autres méthodes

Le Two Pointers Pattern est une technique efficace pour résoudre certains types de problèmes, mais il est important de comprendre ses avantages et ses inconvénients par rapport à d'autres approches. Voici une comparaison avec quelques méthodes alternatives :

### 5.1 Recherche par force brute

**Méthode** : Examiner toutes les combinaisons possibles.

**Comparaison** :
- **Avantage du Two Pointers** : O(n) contre O(n^2) pour de nombreux problèmes.
- **Inconvénient du Two Pointers** : Nécessite souvent que les données soient triées ou structurées d'une certaine manière.

### 5.2 Utilisation de structures de données supplémentaires

**Méthode** : Utiliser des hashsets, des maps ou d'autres structures de données pour stocker des informations intermédiaires.

**Comparaison** :
- **Avantage du Two Pointers** : Généralement O(1) en espace contre O(n) pour les structures supplémentaires.
- **Inconvénient du Two Pointers** : Peut être moins flexible pour certains types de problèmes complexes.

### 5.3 Algorithmes de division et conquête

**Méthode** : Diviser le problème en sous-problèmes plus petits, les résoudre, puis combiner les solutions.

**Comparaison** :
- **Avantage du Two Pointers** : Souvent plus simple à implémenter et à comprendre.
- **Inconvénient du Two Pointers** : Peut être moins efficace pour certains problèmes qui bénéficient d'une approche récursive.

### 5.4 Programmation dynamique

**Méthode** : Résoudre des sous-problèmes et mémoriser leurs solutions pour éviter les calculs redondants.

**Comparaison** :
- **Avantage du Two Pointers** : Généralement moins gourmand en mémoire.
- **Inconvénient du Two Pointers** : Moins adapté aux problèmes d'optimisation complexes.

Voici un tableau récapitulatif des comparaisons :

| Méthode | Complexité temporelle typique | Complexité spatiale | Facilité d'implémentation | Flexibilité |
|---------|--------------------------------|---------------------|---------------------------|-------------|
| Two Pointers | O(n) | O(1) | Élevée | Moyenne |
| Force brute | O(n^2) ou pire | O(1) | Très élevée | Élevée |
| Structures supplémentaires | O(n) | O(n) | Moyenne | Élevée |
| Division et conquête | O(n log n) | O(log n) (pile d'appel) | Moyenne | Élevée |
| Programmation dynamique | Varie | O(n) typiquement | Faible | Très élevée |

Le choix de la méthode dépend donc des caractéristiques spécifiques du problème à résoudre, des contraintes de performance et de mémoire, ainsi que de la complexité de l'implémentation souhaitée.

## 6. Limites et considérations

Bien que le Two Pointers Pattern soit une technique puissante, il est important de comprendre ses limites et les considérations à prendre en compte lors de son utilisation.

### 6.1 Nécessité de données triées ou structurées

- **Limitation** : De nombreuses applications du Two Pointers Pattern nécessitent que les données soient triées ou structurées d'une certaine manière.
- **Considération** : Le coût du tri préalable doit être pris en compte dans l'analyse de la complexité globale.

### 6.2 Applicabilité limitée

- **Limitation** : Le Two Pointers Pattern n'est pas adapté à tous les types de problèmes, en particulier ceux qui nécessitent une vue globale des données.
- **Considération** : Il est important d'évaluer si le problème peut être résolu efficacement avec cette approche avant de l'appliquer.

### 6.3 Gestion des cas particuliers

- **Défi** : Certains problèmes peuvent avoir des cas particuliers qui compliquent l'implémentation du Two Pointers Pattern.
- **Solution** : Une attention particulière doit être portée à la gestion de ces cas lors de la conception de l'algorithme.

### 6.4 Complexité dans les structures de données non linéaires

- **Limitation** : Le Two Pointers Pattern est principalement conçu pour des structures linéaires comme les tableaux ou les chaînes.
- **Adaptation** : Son application à des structures non linéaires (comme les arbres ou les graphes) peut nécessiter des modifications significatives.

Voici un exemple illustrant la gestion d'un cas particulier dans le problème de la suppression des doublons d'un tableau trié :

```python
def remove_duplicates(nums):
    if not nums:
        return 0
    
    # Pointeur pour la position de l'élément unique courant
    unique_pos = 0
    
    for i in range(1, len(nums)):
        if nums[i] != nums[unique_pos]:
            unique_pos += 1
            nums[unique_pos] = nums[i]
    
    # Retourne la longueur du tableau sans doublons
    return unique_pos + 1

# Test
nums = [0,0,1,1,1,2,2,3,3,4]
length = remove_duplicates(nums)
print(f"Longueur du tableau sans doublons: {length}")
print(f"Tableau modifié: {nums[:length]}")
```

### 6.5 Difficulté de parallélisation

- **Limitation** : Les algorithmes basés sur le Two Pointers Pattern sont souvent séquentiels par nature, ce qui peut limiter les possibilités de parallélisation.
- **Considération** : Pour les problèmes nécessitant un traitement hautement parallèle, d'autres approches peuvent être plus appropriées.

### 6.6 Complexité de débogage

- **Défi** : Le débogage des algorithmes utilisant le Two Pointers Pattern peut être complexe, en particulier lorsque les pointeurs se déplacent de manière non intuitive.
- **Solution** : Une bonne pratique consiste à utiliser des assertions et des points de contrôle pour vérifier l'état des pointeurs à des moments clés de l'exécution.

### 7.2 Solutions

```python
# Exercice 1 : Trouver la paire la plus proche d'une somme donnée
def closest_pair_sum(arr, target_sum):
    left, right = 0, len(arr) - 1
    closest_pair = None
    min_diff = float('inf')
    
    while left < right:
        current_sum = arr[left] + arr[right]
        diff = abs(current_sum - target_sum)
        
        if diff < min_diff:
            min_diff = diff
            closest_pair = (arr[left], arr[right])
        
        if current_sum < target_sum:
            left += 1
        else:
            right -= 1
    
    return closest_pair

# Test
print("Exercice 1:", closest_pair_sum([10, 22, 28, 29, 30, 40], 54))
```

```python
# Exercice 2 : Déplacer tous les zéros à la fin
def move_zeros(nums):
    non_zero = 0
    for i in range(len(nums)):
        if nums[i] != 0:
            nums[non_zero], nums[i] = nums[i], nums[non_zero]
            non_zero += 1
    return nums

# Test
print("Exercice 2:", move_zeros([0,1,0,3,12]))
```

```python
# Exercice 3 : Trouver le triplet avec une somme donnée
def find_triplet(arr, target_sum):
    for i in range(len(arr) - 2):
        left, right = i + 1, len(arr) - 1
        while left < right:
            current_sum = arr[i] + arr[left] + arr[right]
            if current_sum == target_sum:
                return (arr[i], arr[left], arr[right])
            elif current_sum < target_sum:
                left += 1
            else:
                right -= 1
    return None

# Tests
print("Exercice 3:", find_triplet([1, 4, 45, 6, 10, 8], 22))
```
